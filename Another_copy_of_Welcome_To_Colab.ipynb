```
# step3:- handling missing data and nan values
#scenario-1:missing values from numeric values
#scenario-2:missing values from non numeric values
#scenario-3:missing values in huge data set

import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])

```

```
X
```

```
# step:4-handling the catagorial data
# which converts the text data into numeric data
from sklearn.preprocessing import OneHotEncoder
#is a class which was imported from S1.preprocessing library
#is used to convert
from sklearn.compose import ColumnTransformer
#  is a class which was imported from
#is used to apply changes of numeric data from catagorial data

#Alexa-0,0,1
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#is a method which converts to txt to binary
X=ct.fit_transform(X)
```

```
X
```

```
#one hot encoder is apply for only input colums
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```
y
```

```
#step5:-split the data into training and testing data
from sklearn.model_selection import train_test_split
#is a class which was import from model_selection
#spliting the data into training and testing
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```
#step6:-feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform (X_test[:,3:])
```

```
#step7:-apply algorithm to  the training data for model
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```
y_pred=classifier.predict(X_test)
```
